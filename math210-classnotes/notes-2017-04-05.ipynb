```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
tax = pd.read_csv("http://www.math.ubc.ca/~pwalls/data/FEDS-2014.csv",encoding='latin',usecols=[0,1,2,3,4,5,6])
```

```python
tax.head()
```

```python
tax.info()
```

```python
tax['Prov/Terr'].unique()
```

```python
tax = tax[tax['Prov/Terr'] != 'TOTAL']
```

```python
tax['Prov/Terr'].unique()
```

Which province has the highest total income?

```python
tax.groupby('Prov/Terr')['Total Income'].sum().sort_values().plot(kind='bar')
```

```python
tax.plot(kind='scatter',x='Total Income',y='Total')
```

```python
tax[tax['Total Income'] > 1e10]
```

```python
tax['Province'] = tax['Prov/Terr'].map(lambda code : int(code[0]))
```

```python
tax.head()
```

```python
tax.plot(kind='scatter',x='Total Income',y='Total',c='Province',figsize=(15,10),colormap='Dark2',s=50,alpha=0.7)
```

## 1. NumPy arrays

### Array Operations

We can do arithmetic operations on arrays like addition, multiplication and division which are performed elementwise.

Array addition/subtraction is the same as vector addition/subtraction:

```python
v = np.array([1,2,3])
w = np.array([1,0,-1])
```

```python
v + w
```

```python
v - w
```

In the same way, array multiplication/division is performed element by element:

```python
v * w
```

```python
w / v
```

Same with exponent operator:

```python
v ** 2
```

```python
# v = [1,2,3]
# w = [1,0,-1]
# v ** w = [1**1, 2**0, 3**(-1)] = [1,1,1/3]
v ** w
```

What?!! It turns out that NumPy arrays homogeneous in the sense that they only carry one type! Since defined arrays with integer entries and most of the entries in `v ** w` are integers, the array remains an integer array and floats are converted to integers.

But note that some operations will change the type of all the entries:

```python
1 / v
```

```python
v.dtype # The entries of v are 64 bit integers
```

```python
(1 / v).dtype # The entries of 1/v are floats
```

This kind of thing doesn't happen often since we're usually dealing with floats all the time!

```python
v = np.arange(0,1,0.1)
```

```python
v
```

```python
w = np.arange(1,11)
```

```python
w
```

```python
v ** w
```

Let's use array operations to plot $y = 1 - x^2$ for $x \in[-2,2]$:

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-2,2,100)
y = 1 - x**2
plt.plot(x,y,'r*');
```

### Vectorized mathematical functions

A vectorized function is just a function which gets applied to every entry of an array.

```python
np.pi
```

```python
x = np.array([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
print(x)
```

```python
np.sin(x)
```

```python
x = np.linspace(-50,50,500)
y = np.sin(x)
plt.plot(x,y,'b.'), plt.ylim([-2,2]), plt.xlim([-50,50]);
```

Let's plot $y = e^{-x^2/10} \cos(10x)$ for $x \in [-10,10]$:

```python
x = np.arange(-10,10,0.01)
y = np.exp(-x**2 / 10) * np.cos(10*x)
plt.plot(x,y), plt.ylim([-2,2]);
```

Let's plot $x = \cos(t)$ and $y = \sin(t)$ for $t \in [0,2\pi]$:

```python
t = np.linspace(0,2*np.pi,1000)
x = np.cos(t)
y = np.sin(t)
plt.plot(x,y), plt.axis('equal'), plt.xlim([-2,2]), plt.ylim([-2,2]);
```

Let's plot $x = t \cos(10t)$ and $y = t \sin(10t)$ for $t \in [0,2\pi]$:

```python
t = np.linspace(0,2*np.pi,1000)
x = t * np.cos(10*t)
y = t * np.sin(10*t)
plt.plot(x,y), plt.axis('equal'), plt.xlim([-2,2]), plt.ylim([-2,2]);
```

```python
t = np.linspace(0,2*np.pi,1000)
x = t * np.cos(10*t)
y = t * np.sin(10*t)
x1 = t * np.cos(8*t)
y1 = t * np.sin(8*t)
plt.plot(x,y,x1,y1,'r'), plt.axis('equal'), plt.xlim([-2,2]), plt.ylim([-2,2]);
```

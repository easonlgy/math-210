## 1. Exploring `pyplot` commands

We use [matplotlib](http://matplotlib.org) to generate plots and we use the [`pyplot` library](http://matplotlib.org/api/pyplot_summary.html) of commands to customize our figures. There are many `pyplot` commands such as:

* `xlim` and `ylim`
* `grid`
* `title`
* `xlabel` and `ylabel`

Let's plot $y = \sin(x) + \sin(4x)$ for $x \in [0,2\pi]$.

```python
x = np.linspace(0,2*np.pi,1000)
y = np.sin(x) + np.sin(4*x)
plt.plot(x,y,'r-.',linewidth=2)
plt.xlim([0,2*np.pi]), plt.ylim([-3,3]), plt.grid('on',linestyle='-.')
plt.title('Sum pf sine waves'), plt.xlabel('Time (s)'), plt.ylabel('Amplitude (mm)');
```

There is also a special command called `show`. We use this when we want to render the current figure. Jupyter notebook calls this for the current figure when it reaches the end of a cell. So we will use this when we want to render several figures from within the same cell.

```python
x = np.linspace(0,2*np.pi,1000)
y1 = np.sin(x) + np.sin(4*x)
plt.plot(x,y1)
plt.show()

y2 = np.sin(x) + np.sin(4*x) + np.sin(4*x)
plt.plot(x,y2,'r')
plt.show()
```

We can also generate a legend for a figure with several lines. For example, let's plot $y = \sin(x)$ and $y = \cos(x)$ in the same figure with a legend.

```python
t = np.linspace(0,4*np.pi,100)
y1 = np.sin(t)
y2 = np.cos(t)
plt.plot(t,y1,t,y2), plt.xlim([0,4*np.pi]), plt.ylim([-2,2]) 
plt.legend(['y = sin(t)','y = cos(t)']);
```

## 2. Subplot

The command `subplot` takes at least 3 inputs $n$, $m$ and $i$ and creates a figure with a $n$ by $m$ grid of subplots and then sets the $i$th subplot (counting across the rows) as the current plot (ie. current axes object).

```python
for k in range(1,5):
    plt.subplot(2,2,k)
    x = np.linspace(-5,5,1000)
    y = x**(k-1) * np.arctan(x) / (1 + x**2)
    plt.plot(x,y), plt.grid('on'), plt.xlim([-5,5])
    plt.title('k = ' + str(k)), plt.tight_layout()
```

## 3. Different kinds of plots

So far we have just been creating line plots. Let's create some other kinds.

### Scatter plots

```python
N = 10000
x = np.random.randn(N)
y = np.random.rand(N)
plt.scatter(x,y,lw=0,alpha=0.2,c='green');
```

### Bar plots

```python
x = np.arange(0,10)
y = np.random.rand(10)
plt.bar(x,y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
# Import data
elections = pd.read_csv("http://www.math.ubc.ca/~pwalls/data/elections.csv",encoding='latin')
```

## 1.

```python
# Specify elected candidates
elected = elections[elections["Elected"] == 1]
# Create a new column which combine the 'Last Name' and 'First Name'
elected['candidates'] = elected['Last Name'] + elected['First Name']
```

```python
# Calculate population of each occupation
electedoccupations = elected.groupby('Occupation',sort=False).size()
# Sort the result by descending number
electedoccupations.sort_values(ascending=False)
```

### The most common occupation of elected members of parliament is lawyer (2046 people).

## 2.

```python
# Find all parties which had people got elected and then get rid of the repeated result 
partyelected = elected['Party'].unique()
# Give the size of the parties which had people got elected
partyelected.size
```

### There are 46 different parties have had a candidate elected.

## 3.

```python
# Specify people who got 100% votes
hundredpercent = elections[elections['Votes (%)'] == 100]
hundredpercent
```

### Duncan H. McAlister and Hon. Sydney Chilton Mewburn received 100% of the votes in an election.

## 4.

```python
# From elected people, calculate the times each one got elected
electedtimes = elected.groupby('candidates',sort=False).size()
# Sort the result by descending number
electedtimes.sort_values(ascending=False)
```

### DIEFENBAKER John George and Gray Herb have been elected the most times (13 times).

```python
# Specify not elected candidates
notelected = elections[elections['Elected'] == 0]
# Create a new column which combines the 'Last Name' and 'First Name'
notelected['candidates'] = notelected['Last Name'] + notelected['First Name']
```

```python
# From not elected people, calculate the times each one not got elected
notelectedtimes = notelected.groupby('candidates',sort=False).size()
# Sort the result by descending number
notelectedtimes.sort_values(ascending=False)
```

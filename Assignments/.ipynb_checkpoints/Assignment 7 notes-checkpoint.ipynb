```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.integrate as spi
```

## 1.

This is a 2D system of second order ODEs. To rewrite this as a first order system, let's introduce variable $u_0 = x$, $u_1 = x'$

$$
\begin{align}
u_0' = u_1 \\
u_1' = \mu (1-u_0^2)u_1 - u_0
\end{align}
$$

```python
def van_der_pol(mu,x0,tf):
    def f(u,t):
        dudt = [0,0]
        dudt[0] = u[1]
        dudt[1] = mu * (1 - u[0]**2) * u[1] - u[0]
        return dudt
    initial_u = x0
    t = np.linspace(0,tf,10000)
    U = spi.odeint(f,initial_u,t)
    plt.plot(U[:,0],U[:,1])
```

```python
van_der_pol(10,[1,0],200)
```

## 2.

This is a 2D system of second order ODEs. To rewrite this as a first order system, let's introduce variable $u_0 = \theta$, $u_1 = \theta'$

$$
\begin{align}
u_0' = u_1 \\
u_1' = -\frac{g}{L} u_0
\end{align}
$$

```python
def pendulum(L,theta0,tf):
    g = 9.81
    def f(u,t):
        dudt = [0,0]
        dudt[0] = u[1]
        dudt[1] = - g/L * u[0]
        return dudt
    initial_u = theta0
    t = np.linspace(0,tf,10000)
    U = spi.odeint(f,initial_u,t)
    plt.plot(t,U[:,0])
```

```python
pendulum(100,[1,0],200)
```

## 3.

This is a 4D system of second order ODEs. To rewrite this as a first order system, let's introduce variable $u_0 = x$, $u_1 = x'$,$u_2 = y$,$u_3 = y'$

$$
\begin{align}
u_0' = u_1 \\
u_1' = -\omega^2 u_0 + 2\Omega u_3 \sin(\phi) \\
u_2' = u_3 \\
u_3' = -\omega^2 u_2 - 2\Omega u_1 \sin(\phi)
\end{align}
$$

```python
def coriolis(w,Omega,phi,initial,tf):
    def f(u,t):
        dudt = [0,0,0,0]
        dudt[0] = u[1]
        dudt[1] = -w**2 * u[0] + 2*Omega*u[3]*np.sin(phi)
        dudt[2] = u[3]
        dudt[3] = -w**2 * u[2] - 2*Omega*u[1]*np.sin(phi)
        return dudt
    initial_u = initial
    
    t = np.linspace(0,tf,10000)
    U = spi.odeint(f,initial_u,t)
    plt.plot(U[:,0],U[:,3])
```

```python
coriolis(10,10,np.pi,[0,1,0,1],200)
```

## 4.

```python
def trig_ode(alpha=1,beta=1,y0=0,tf=10):
    def f(y,t):
        return np.sin(alpha*y) + np.cos(beta*t)
    t = np.linspace(0,tf,10000)
    y = spi.odeint(f,y0,t)
    plt.plot(t,y)
```

```python
trig_ode(10,10,0,10)
```

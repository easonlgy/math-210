### 6. Some other economics patterns

#### CPI

Let's see some other economics patterns, begin with the CPI. [CPI](https://en.wikipedia.org/wiki/Consumer_price_index), a.k.a. the consumer price index, measures changes in the price level of market basket of consumer goods and services purchased by households. The annual percentage change in a CPI is used as a measure of inflation.

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111)
ax.plot(CPI.index,CPI['China'])
ax.plot(CPI.index,CPI['India'])
ax.plot(CPI.index,CPI['Japan'])
ax.plot(CPI.index,CPI['United States'])
ax.legend()
ax.xaxis.set_ticks(CPI.index)
plt.tight_layout()
plt.xticks(rotation=90);
```

According to the definition, inflation rate is

$$
\frac{2^{nd}year CPI - 1^{st}year CPI}{1^{st}year CPI} 100
$$

which is just the slope of the cureve, i.e. the steeper the slope, the higher inflation rate. We can see that India has the highest inflation rate while China has the almost same level inflation rate with Japan and United States.

### 7. How soon will China become a developed country

People who think China is a developed country has their proof that China has the world's second largest GDP, following the United States. Then I will say China has low GDP per capita. And they argue back that because of the huge population. I agree with that. 

However, nowadays in China, primary and secondary sectors take up almost 60% (10% & 45% seperately) of the total GDP. China is known as low-cost labor, so that the mechanization level is just middle class thus with huge labor input. If we cut the population, the GDP will also have a drop. China has already been on the road to switch economy sector, but it takes time.

#### To chase the world

Let's calculate when will China reach the world's average under some assumptions:
  1. The population growth rate stablized at about 0.5% last decade, so we assume it will continue at the same level
  2. The effect of the "One Child Policy" will occur in 70 years (around 2049), causing a huge decline in population, let's assume it will not happen before 2049 and we will first calculate what will happen before 2049
  3. The GDP growth rate stablized at about 7% level
  4. First three assumptions are all about China
  5. Assume World GDP and population growth rate at 2.7% and 1.2% with the same reasons above

```python
n = 0
CHNGDP = GDP['China'].loc[2015]
CHNPPL = PPL['China'].loc[2015]
WORLDGDP = GDP['World'].loc[2015]
WORLDPPL = PPL['World'].loc[2015]
CHNGDPPC = CHNGDP / CHNPPL
WORLDGDPPC = WORLDGDP / WORLDPPL

while CHNGDPPC < WORLDGDPPC:
    CHNGDP = CHNGDP * (1+0.07)
    CHNPPL = CHNPPL * (1+0.005)
    WORLDGDP = WORLDGDP * (1+0.027)
    WORLDPPL = WORLDPPL * (1+0.012)
    CHNGDPPC = CHNGDP / CHNPPL
    WORLDGDPPC = WORLDGDP / WORLDPPL
    n += 1
n
```

Well, that is interesting! China can reach world average GDP per capita in 5 years which is year 2020. Let's see when will China catch up those developed countries.

#### To chase the America

We also need some assumptions:

  1. The assumptions for China as above
  2. The GDP growth rate of United States stables at around 2.5%
  3. The Population growth rate of United States stables at around 0.78%

```python
n = 0
CHNGDP = GDP['China'].loc[2015]
CHNPPL = PPL['China'].loc[2015]
USGDP = GDP['United States'].loc[2015]
USPPL = PPL['United States'].loc[2015]
CHNGDPPC = CHNGDP / CHNPPL
USGDPPPC = USGDP / USPPL

while CHNGDPPC < USGDPPPC:
    CHNGDP = CHNGDP * (1+0.07)
    CHNPPL = CHNPPL * (1+0.005)
    USGDP = USGDP * (1+0.025)
    USPPL = USPPL * (1+0.0078)
    CHNGDPPC = CHNGDP / CHNPPL
    USGDPPPC = USGDP / USPPL
    n += 1
n
```

43 means China can catch up United States' GDP per capita in 43 years, which is year 2057. That is beyond our assumption year 2049, so we can not predit when there comes a huge population decline, whether China has switched the economy sector by then or not. We can just use 43 years as a reference.
